```python
import time
from selenium import webdriver
import pandas as pd
driver = webdriver.Firefox(executable_path="C:/Users/farzadali/Documents/Python Scripts/geckodriver.exe")

data= pd.read_csv('input/Import_HS_Codes_v5.csv', dtype=str)
Import_HS_Codes = data['Import_HS_Codes']
print(len(Import_HS_Codes))
url= "https://www.weboc.gov.pk/(S(ck3swvgh1b3a0voixh4d5qcs))/Shared/TariffList.aspx"
driver.get(url)
```

```python

l=[]

#for i in range(0, 50):
for hsCode in Import_HS_Codes:
    d={}

    #print(hsCode)
    
    try:
        
        inputField= driver.find_element_by_xpath('//*[@id="TariffList_txtSearch"]')
        inputField.send_keys(str(hsCode)) #enter HS Code
        #inputField.send_keys(Import_HS_Codes[i]) #enter HS Code
        
        time.sleep(4)
        search_button = driver.find_element_by_xpath('//*[@id="TariffList_btnSearch"]')
        search_button.click()
        time.sleep(4)



        d["HS_Code"] = hsCode


        try:
            row_4_element = driver.find_element_by_xpath('/html/body/form/div[3]/div[2]/table[2]/tbody/tr[4]/td[1]')
            d["Row_4"] = row_4_element.text
        except:
            d["Row_4"] = None
            
        try:
            row_5_element = driver.find_element_by_xpath('/html/body/form/div[3]/div[2]/table[2]/tbody/tr[5]/td[1]')
            d["Row_5"] = row_5_element.text
        except:
            d["Row_5"] = None
            
        try:
            row_6_element = driver.find_element_by_xpath('/html/body/form/div[3]/div[2]/table[2]/tbody/tr[6]/td[1]')
            d["Row_6"] = row_6_element.text
        except:
            d["Row_6"] = None

        clear_button = driver.find_element_by_xpath('//*[@id="TariffList_btnClear"]')
        clear_button.click()
        time.sleep(4)
        
    except:
        pass
#         d["HS_Code"] = hsCode
#         d["Row_4"] = None
#         d["Row_5"] = None
#         d["Row_6"] = None
        

#         clear_button = driver.find_element_by_xpath('//*[@id="TariffList_btnClear"]')
#         clear_button.click()
    time.sleep(4)
    print(d)
    
    l.append(d)
```

```python
print(len(l))

# Convert the list to a dataframe
df = pd.DataFrame(l)

# Save the dataframe as a CSV file
df.to_csv('output_v3.csv', index=False)

```
